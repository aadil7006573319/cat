Question 1 : Define Data Transformation in ETL and explain why it is important.
  -### Data Transformation in ETL

**Data Transformation** is the **T (Transform)** phase of the **ETL (Extract, Transform, Load)** process. It involves converting, cleaning, and modifying extracted data into a suitable format before loading it into a target system such as a data warehouse, database, or analytics platform.

**ETL Process:**

1. **Extract** – Collect data from various sources (databases, files, APIs, etc.).
2. **Transform** – Clean, validate, standardize, aggregate, and convert the data.
3. **Load** – Store the transformed data in the target system.

### Examples of Data Transformation

* Removing duplicate records.
* Correcting missing or invalid values.
* Converting data types (e.g., text to date format).
* Standardizing formats (e.g., changing all dates to YYYY-MM-DD).
* Combining data from multiple sources.
* Calculating new fields such as total sales or average revenue.
* Filtering unnecessary data.

### Importance of Data Transformation

1. **Improves Data Quality**

   * Cleans inaccurate, incomplete, or duplicate data.
   * Ensures reliable and consistent information.

2. **Ensures Data Consistency**

   * Standardizes data formats and naming conventions across different sources.

3. **Enhances Decision-Making**

   * Provides accurate and meaningful data for reporting, business intelligence, and analytics.

4. **Facilitates Data Integration**

   * Combines data from multiple systems into a unified format.

5. **Increases Operational Efficiency**

   * Automates data preparation tasks, reducing manual effort and errors.

6. **Supports Business Rules**

   * Applies organizational policies and calculations before data is stored.

7. **Optimizes Performance**

   * Aggregated and structured data improves query and reporting performance.

### Conclusion

Data Transformation is a critical step in ETL that converts raw data into a clean, consistent, and usable format. It ensures high-quality data, supports accurate analysis, and enables organizations to make informed business decisions.

  Question 4 : A dataset has missing values in the “Age” column. Suggest two techniques to handle this and
explain when they should be used.

   ### Question 4: A dataset has missing values in the “Age” column. Suggest two techniques to handle this and explain when they should be used.

Missing values in the **Age** column can be handled using several techniques. Two common methods are:

#### 1. Mean/Median Imputation

* Replace missing age values with the **mean** or **median** age of the dataset.
* **When to use:**

  * Use the **mean** when the age data is normally distributed and has few outliers.
  * Use the **median** when the data contains outliers or is skewed, as it is less affected by extreme values.

**Example:**
If the ages are 20, 22, 25, 27, and a value is missing:

* Mean age = 23.5 → replace missing value with 23.5.
* Median age = 23.5 → replace missing value with 23.5.

#### 2. Record Deletion (Listwise Deletion)

* Remove rows that contain missing age values.
* **When to use:**

  * When the number of missing values is very small compared to the total dataset.
  * When deleting those records will not significantly affect the analysis or introduce bias.

### Conclusion

* **Mean/Median Imputation** is useful when you want to retain all records and maintain dataset size.
* **Record Deletion** is appropriate when only a few values are missing and their removal has minimal impact on the results.

   ### Question 6: What is One-Hot Encoding? Give an Example with the Categories: “Red, Blue, Green”

**One-Hot Encoding** is a data preprocessing technique used to convert **categorical data** into a numerical format that machine learning algorithms can understand. Each category is represented by a separate binary (0 or 1) column.

### Example

Suppose we have a feature **Color** with three categories:

```text
Red, Blue, Green
```

After One-Hot Encoding:

| Color | Red | Blue | Green |
| ----- | --- | ---- | ----- |
| Red   | 1   | 0    | 0     |
| Blue  | 0   | 1    | 0     |
| Green | 0   | 0    | 1     |

### Encoded Values

```text
Red   → [1, 0, 0]
Blue  → [0, 1, 0]
Green → [0, 0, 1]
```

### Why Use One-Hot Encoding?

* Converts categorical values into numerical form.
* Prevents algorithms from assuming an order among categories.
* Improves the performance of many machine learning models.

### Conclusion

One-Hot Encoding creates a separate binary column for each category. For the categories **Red, Blue, Green**, the encoded vectors are:

* Red → [1, 0, 0]
* Blue → [0, 1, 0]
* Green → [0, 0, 1]


  Question 7 : Explain the difference between Data Integration and Data Mapping in ETL.


    ### Question 7: Explain the Difference Between Data Integration and Data Mapping in ETL

| **Data Integration**                                                                         | **Data Mapping**                                                                               |
| -------------------------------------------------------------------------------------------- | ---------------------------------------------------------------------------------------------- |
| The process of combining data from multiple sources into a unified view.                     | The process of defining how data fields from a source correspond to fields in a target system. |
| Focuses on merging and consolidating data.                                                   | Focuses on matching and transforming individual data elements.                                 |
| Ensures data from different systems can be used together.                                    | Ensures data is transferred correctly between systems.                                         |
| Involves activities such as data extraction, transformation, and consolidation.              | Involves creating relationships between source and target fields.                              |
| Example: Combining customer data from CRM, sales, and billing systems into a data warehouse. | Example: Mapping the source field “Cust_Name” to the target field “Customer_Name”.             |

### Data Integration

Data Integration is the process of collecting and combining data from different sources to create a single, consistent dataset for reporting, analysis, or storage.

**Example:**

* Customer data from a CRM system
* Sales data from an ERP system
* Website data from analytics tools

These datasets are integrated into one data warehouse.

### Data Mapping

Data Mapping is the process of defining how source data fields correspond to destination fields during ETL.

**Example:**

| Source Field | Target Field  |
| ------------ | ------------- |
| Cust_ID      | Customer_ID   |
| Cust_Name    | Customer_Name |
| DOB          | Date_of_Birth |

### Conclusion

* **Data Integration** combines data from multiple sources into a unified dataset.
* **Data Mapping** defines how individual fields from source systems are matched and transformed before loading into the target system.
* Data Mapping is an important step within the broader Data Integration process.

    ### Question 8: Explain Why Z-score Standardization is Preferred Over Min-Max Scaling When Outliers Exist

**Z-score Standardization** is often preferred over **Min-Max Scaling** when a dataset contains outliers because it is less sensitive to extreme values.

### Min-Max Scaling

Min-Max Scaling transforms data into a fixed range, usually 0 to 1:

X' = \frac{X - X_{\min}}{X_{\max} - X_{\min}}

Since it uses the **minimum** and **maximum** values, a single outlier can significantly affect the scaling. As a result, most data points may become compressed into a small range.

**Example:**
If ages are:

```text
20, 22, 25, 27, 100
```

The value 100 is an outlier. Min-Max Scaling uses 20 and 100 as the range, causing the other values to be clustered very close together.

### Z-score Standardization

Z-score Standardization transforms data based on the mean and standard deviation:

It measures how far each value is from the mean in terms of standard deviations. Although outliers affect the mean and standard deviation, their impact is generally less severe than in Min-Max Scaling.

### Why Z-score is Preferred with Outliers

1. **Less dependent on extreme minimum and maximum values.**
2. **Preserves the distribution of the data better.**
3. **Prevents normal observations from being squeezed into a narrow range.**
4. **Works well with many machine learning algorithms that assume standardized data.**

### Conclusion

When outliers are present, **Z-score Standardization** is usually preferred because it is less affected by extreme values, while **Min-Max Scaling** can distort the scaled data by relying heavily on the dataset's minimum and maximum values.
